```python
for i in range(data.shape[0]):
    phones = data.iloc[i, 7]
    len_phone = len(data.iloc[i, 7])
    number_frm = data.iloc[i, 6]
    frames_for_each_phone = data.iloc[i, 6]//len(data.iloc[i, 7])
    initial_frame = data.iloc[i, 4]
    print(initial_frame)
    frames_left= number_frm
    print(len(data.iloc[i, 7]),'---->',data.iloc[i, 6],'------->',data.iloc[i, 6]/len(data.iloc[i, 7]))
    frames_to_allot = number_frm
    for pho in range(len(phones)):
        if pho == range(len(phones))[-1]: # for last part of phoneme only
            print(phones[pho],'----->',frames_to_allot)
            frames_to_allot-=frames_for_each_phone
            print(phones[pho],'-----frame number--->',initial_frame,'----to---',initial_frame+frames_left)
        else:
            print(phones[pho],'----->',frames_for_each_phone)
            frames_to_allot-=frames_for_each_phone
            print(phones[pho],'-----frame number--->',initial_frame,'----to---',initial_frame+frames_for_each_phone)
            initial_frame=initial_frame+frames_for_each_phone
            frames_left -= frames_for_each_phone
            
    print('----------------------------------------------')
```

```python
phones_list=[]
init_range=[]
final_range=[]
for i in range(data.shape[0]):
    phones = data.iloc[i, 7]
    len_phone = len(data.iloc[i, 7])
    number_frm = data.iloc[i, 6]
    frames_for_each_phone = data.iloc[i, 6]//len(data.iloc[i, 7])
    initial_frame = data.iloc[i, 4]
    print(initial_frame)
    frames_left= number_frm
    print(len(data.iloc[i, 7]),'---->',data.iloc[i, 6],'------->',data.iloc[i, 6]/len(data.iloc[i, 7]))
    frames_to_allot = number_frm
    for pho in range(len(phones)):
        if pho == range(len(phones))[-1]: # for last part of phoneme only
            print(phones[pho],'----->',frames_to_allot)
            frames_to_allot-=frames_for_each_phone
            print(phones[pho],'-----frame number--->',initial_frame,'----to---',initial_frame+frames_left)
            phones_list.append(phones[pho])
            init_range.append(initial_frame)
            final_range.append(initial_frame+frames_left)
            
        else:
            print(phones[pho],'----->',frames_for_each_phone)
            frames_to_allot-=frames_for_each_phone
            print(phones[pho],'-----frame number--->',initial_frame,'----to---',initial_frame+frames_for_each_phone)
            phones_list.append(phones[pho])
            init_range.append(initial_frame)
            final_range.append(initial_frame+frames_for_each_phone)
            
            initial_frame=initial_frame+frames_for_each_phone
            frames_left -= frames_for_each_phone
         
    print('----------------------------------------------')
data_new= pd.DataFrame(columns= ['Phonemes','Init','Final'])
data_new['Phonemes']=phones_list
data_new['Init']=init_range
data_new['Final']=final_range
```

```python
data_new.to_csv('final.csv')
```
